```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

np.random.seed(0)

age =np.random.randint(20, 60, size=500) #500 data points
salary =age * 500 + np.random.randint(10000, 40000, size=500)

data ={
    "Age" : age,
    "Salary" :salary
}

df =pd.DataFrame(data)
df.head()
```

```python
x = df[['Age']]
y = df[['Salary']]

plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x="Age", y="Salary")
plt.show()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

Lr = LinearRegression()
Lr.fit(x_train, y_train)

y_pred = Lr.predict(x_test)

mae = mean_absolute_error(y_test, y_pred) 
mse = mean_squared_error(y_test, y_pred) 
r2 = r2_score(y_test, y_pred) 

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}") 
print(f"R² Score: {r2}")

plt.figure(figsize=(10,8))
sns.scatterplot(data=df, x="Age", y="Salary", label="Actual")
sns.scatterplot(x=x_test['Age'], y=y_pred.flatten(), label="Predicted", color='r')
plt.legend()
plt.show()
```

```python
R² Score: Closer to 1 is better.

MAE and MSE: Lower values are better.
```
